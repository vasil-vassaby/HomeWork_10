f(x) = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30

    Определить корни

    Найти интервалы, на которых функция возрастает

    Найти интервалы, на которых функция убывает

    Построить график

    Вычислить вершину

    Определить промежутки, на котором f > 0

    Определить промежутки, на котором f < 0


```python
# импортируем необходимые модули:
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve
```

```python
# определяем функцию для заданного уравнения:
def f(x):
    return a * x**4 * np.sin(np.cos(x)) + b * x**3 + c * x**2 + d * x + e
```

```python
# вводим значения переменных и для исключения бесконечности определяемых значений вводим диапазон по х:
a, b, c, d, e = -12, -18, 5, 10, -30
x = np.arange(-2, 2.5, 0.01)
```

```python
# находим корни уравнения на отрезке х = от -2.5 до 2.5:
x1 = fsolve(f, -2)
x2 = fsolve(f, 2)
```

```python
# определяем функцию для поиска координаты вершин по оси х:
def y(x, min_y):
    return a * x**4 * np.sin(np.cos(x)) + b * x**3 + c * x**2 + d * x + e - min_y
```

```python
# так как на диапазоне от -2.5 до 2.5 две вершины, разобьем его на 2 отрезка, первый х = от -1 до 0, 
# находим значение кординат (min_x, min_y):
x = np.arange(-1, 0, 0.01)
min_y2 = min(f(x))
min_x2 = fsolve(y, -1, min_y2)
```

```python
# второй х = от 1 до 2, так же находим значение кординат (min_x, min_y) для вершины:
x = np.arange(1, 2, 0.01)
min_y1 = min(f(x))
min_x1 = fsolve(y, 1, min_y1)
```

```python
# оформление графика:
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.title(r'$f(x)=-12x^4sin(cos(x))-18x^3+5x^2+10x-30$')
#корни уравнения на отрезке от -2.5 до 2.5:
plt.scatter(x1, 0, color='orange', s=50, marker='o')
plt.scatter(x2, 0, color='orange', s=50, marker='o')
#вершины на отрезке от -2.5 до 2.5:
plt.scatter(min_x1, min_y1, color='red', s=50, marker='x')
plt.text(1.3, -99, f'({round(min_x1[0], 2)}; {round(min_y1, 2)})')

plt.scatter(min_x2, min_y2, color='red', s=50, marker='x')
plt.text(-0.9, -70, f'({round(min_x2[0], 2)}; {round(min_y2, 2)})')
# интервал, на котором функция убывает:
grow_down_x = np.arange(-2.5, x1, 0.01)
plt.plot(grow_down_x, f(grow_down_x), color='black', label = 'убывает > 0')
plt.text(-2.2, 200, 'убывает > 0', rotation = -75)
# интервал, на котором функция возрастает:
grow_up_x = np.arange(x2, 3, 0.01)
plt.plot(grow_up_x, f(grow_up_x), color='green', label = 'возрастает > 0')
plt.text(2.2, 80, 'возрастает > 0', rotation = 70)
# интервал, на котором функция убывает:
grow_down_x1 = np.arange(x1, min_x2, 0.01)
plt.plot(grow_down_x1, f(grow_down_x1), linestyle = '--', color='blue', label = 'убывает < 0')
# интервал, на котором функция возрастает:
grow_up_x1 = np.arange(min_x1, x2, 0.01)
plt.plot(grow_up_x1, f(grow_up_x1), linestyle = '--', color='crimson', label = 'возрастает < 0')
# интервал между вершинами
x0 = np.arange(min_x2, min_x1, 0.01)
plt.plot(x0, f(x0), linestyle = '--', color='darkmagenta', label = f'переход < 0')

plt.legend()

plt.grid()
```
